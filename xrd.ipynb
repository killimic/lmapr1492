```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
```

```python
with MPRester("el3corSTO535RH2iv") as m:
# Structure for material id
    structure = m.get_structure_by_material_id("mp-15793")
```

```python
xrdcal = XRDCalculator('CuKa1')
#https://pydoc.net/pymatgen/2017.12.30/pymatgen.analysis.diffraction.xrd/
#Info sur XRDCalculator et notamment type de rayon X incident
patt = xrdcal.get_pattern(structure)
```

```python
print("Premier pic du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å :")
print("\nangle : 2θ(°) = \t",patt.x[0])
print("Intensité du pic : I = \t",patt.y[0])
print("Indices hkl associés : \t",patt.hkls[0])
print("Distance interplanaire : d(Å) = \t",patt.d_hkls[0])
```

```python
print("Deuxième pic du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å :")
print("\nangle : 2θ = \t",patt.x[1])
print("Intensité du pic : I = \t",patt.y[1])
print("Indices hkl associés : \t",patt.hkls[1])
print("Distance interplanaire : d = \t",patt.d_hkls[1])
```

```python
print("Troisième pic du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å :")
print("\nangle : 2θ = \t",patt.x[2])
print("Intensité du pic : I = \t",patt.y[2])
print("Indices hkl associés : \t",patt.hkls[2])
print("Distance interplanaire : d = \t",patt.d_hkls[2])
```

```python
print("Intensités relatives des 3 premiers pics du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å")
xrdcal.show_plot(structure,two_theta_range=(10,27),annotate_peaks=True,ax=None,with_labels=True)
```
